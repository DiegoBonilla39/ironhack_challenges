```python
import pandas as pd

df = pd.read_csv("../data/input/IMDB-Movie-Data.csv")
```

# üêº Challenge 1. Warm up

```python
def movie_bin(votes):
    if votes >= 0 and votes <= 999:
        return 1
    elif votes >= 1000 and votes <= 9999:
        return 2
    elif votes >= 10000 and votes <= 99999:
        return 3
    elif votes >= 100000 and votes <= 999999:
        return 4
    elif votes > 1000000:
        return 5

df["bin"] = df["Votes"].apply(movie_bin)
df.head(5)
```

# üêº üêº üêº Challenge 3. Using the lambda

```python
def new_ranking(genre,bins):
    if "Thriller" in genre:
        return bins + 1
    elif "Comedy" in genre:
        return bins - 1
    else:
        return bins

df['new ranking'] = df.apply(lambda x : new_ranking( x["Genre"], x["bin"] ), axis = 1)
df.head(1)
```

# üêº üêº üêº üêº Challenge 4. Now the real stuff

```python
def prime_number(title,votes):
    num = int(sum([ord(i) for i in title])/votes)
    if num > 1:
        for i in range(2, int(num/2)+1):
            if (num % i) == 0:
                return False
                break
        else:
            return True
    else:
        return False

df['prime number'] = df.apply(lambda x : prime_number( x["Title"], x["Votes"] ), axis = 1)

# Movies that its title divided by their number of votes retrieve a prime number
filtered_df = df[df['prime number'] == True]
filtered_df.head()
```

# üêº üêº üêº üêº üêº Challenge 5. And finally some fantasy

```python
# The "New Ranking 2" will punctuate the movies from 0 to 10 taking into consideration the following weighted variables:

    # Popular opinion (Votes) = Pointing between 0 and 4 from lowest to highes
    # Movie critics opinion (Rating) = Pointing between 0 and 4 from lowest to highest
    # Movie lenght (Runtime (Minutes)) = Pointing between 0 and 2
        # The median value represents 2 points and the minimun and maximun values represent 0 points

def new_ranking2(votes,runtime,rating):
    vo = 0.0
    ru = 0.0
    ra = 0.0
    
    # Value for rating (max 4 points)
    
    if rating == 1.900000:
        ra += 0
    elif rating == 9.000000:
        ra += 4
    else:
        ra += (rating*4)/9.000000
    
    # Value for votes (max 4 points)
    
    if votes == 6.100000e+01:
        vo += 0
    elif votes == 1.791916e+06:
        vo += 4
    else:
        vo += (votes*4)/1.791916e+06
    
    # Value for runtime (max 2 points)
    
    if runtime == 111.000000:
        ru += 2
    elif runtime == 66.000000 or runtime == 191.000000:
        ru += 0
    elif runtime > 111.000000 and runtime < 191.000000:
        ru += 2-(((runtime - 111.000000)*2)/111.000000)
    elif runtime > 66.000000 and runtime < 111.000000:
        ru += (runtime*2)/111.000000
    
    #Final values (max 10 points)
    
    return vo+ru+ra
    
df['new ranking 2'] = df.apply(lambda x : new_ranking2( x["Votes"], x["Runtime (Minutes)"], x["Rating"]), axis = 1)

# Top 5 movies by "New Ranking 2"

df.sort_values(by=['new ranking 2'], ascending=False).head(5)
```

# üêº üêº üêº üêº üêº üêº Challenge 6. Freaky bonus

```python
def freaky_bonus(text,votes,revenue):
    import hashlib
    import string

    m = hashlib.sha256(text.encode('UTF-8'))
    sum_digits = sum([int(i) for i in m.hexdigest() if i in string.digits])

    if votes <= sum_digits <= revenue*10000:
        return True
    else:
        return False

df['freaky bonus'] = df.apply(lambda x : freaky_bonus( x["Description"], x["Votes"], x["Revenue (Millions)"]), axis = 1)


# Movies which the sum of all numeric values of the string description hash (SHA256) 
# are between their revenue and their number of votes

filtered_df2 = df[df['freaky bonus'] == True]
filtered_df2
```
